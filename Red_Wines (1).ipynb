|Feature|Explain|
|-------|-------|
|**fixed acidity**|most acids involved with wine or fixed or nonvolatile|
|**volatile acidity**|the amount of acetic acid in wine|
|**citric acid**|the amount of citric acid in wine|
|**residual sugar**|the amount of sugar remaining after fermentation stops|
|**chlorides**|the amount of salt in the wine.|
|**free sulfur dioxide**|the amount of free sulfur dioxide in the wine(those available to react and thus exhibit both germicidal and antioxidant properties)|
|**total sulfur dioxide**|amount of free and bound forms of SO2|
|**density**|the measurement of how tightly a material is packed together|
|**PH**|describes how acidic or basic a wine is on a scale from 0 (very acidic) to 14 (very basic); most wines are between 3-4|
|**Alcohol**|the percent alcohol content of the wine|
|**quality**|output variable (based on sensory data, score between 3 and 8)|

<a id="norm"></a>
<h2><b><span style="color:#009b2d">üç∑ Step 3.3 | </span><span style="color:#9b006e">Normalization</span></b></h2>

```python
df.corr()
```

```python
df['quality'] = df.quality.apply(lambda x:1 if x>=7 else 0)
```

```python
df['quality'].value_counts()
```

```python
X = df.drop('quality', axis=1)
y = df['quality']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
print("X_train ", X_train.shape)
print("y_train ", y_train.shape)
print("X_test ", X_test.shape)
print("y_test ", y_test.shape)
```

<a id="modeling"></a>
<h1><b><p style="background-image: url(https://i.postimg.cc/C5rCv1td/th1.jpg);font-family:tahoma;font-size:120%;color:white;text-align:center;border-radius:15px 50px; padding:10px; border:solid 4px #09375b; box-shadow: 10px 10px 10px #042b4c">Step 4 | Modeling</p></b></h1>

<a id="rf"></a>
<h2><b><span style="color:#009b2d">üç∑ Step 4.1 | </span><span style="color:#9b006e">Logistic Regression</span></b></h2> 

```python
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
logreg_pred = logreg.predict(X_test)
logreg_acc = accuracy_score(logreg_pred, y_test)
print("Test accuracy is: {:.2f}%".format(logreg_acc*100))
```

```python
print(classification_report(y_test, logreg_pred))
```

<a id="rf"></a>
<h2><b><span style="color:#009b2d">üç∑ Step 4.2 | </span><span style="color:#9b006e">Decision Tree</span></b></h2>  

```python
dtree = DecisionTreeClassifier()
dtree.fit(X_train, y_train)
dtree_pred = dtree.predict(X_test)
dtree_acc = accuracy_score(dtree_pred, y_test)
print("Test accuracy: {:.2f}%".format(dtree_acc*100))
```

```python
print(classification_report(y_test, logreg_pred))
```

<a id="rf"></a>
<h2><b><span style="color:#009b2d">üç∑ Step 4.3 | </span><span style="color:#9b006e">Random Forest</span></b></h2>   

```python
rforest = RandomForestClassifier()
rforest.fit(X_train, y_train)
rforest_pred = rforest.predict(X_test)
rforest_acc = accuracy_score(rforest_pred, y_test)
print("Test accuracy: {:.2f}%".format(rforest_acc*100))
```

```python
print(classification_report(y_test, logreg_pred))
```

<a id="rf"></a>
<h2><b><span style="color:#009b2d">üç∑ Step 4.4 | </span><span style="color:#9b006e">SVC</span></b></h2> 

```python
svc = SVC()
svc.fit(X_train, y_train)
svc_pred = svc.predict(X_test)
svc_acc = accuracy_score(svc_pred, y_test)
print("Test accuracy: {:.2f}%".format(svc_acc*100))
```

```python
print(classification_report(y_test, logreg_pred))
```

<a id="rf"></a>
<h2><b><span style="color:#009b2d">üç∑ Step 4.5 | </span><span style="color:#9b006e">KNeighborsClassifier</span></b></h2> 

```python
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
knn_pred = knn.predict(X_test)
knn_acc = accuracy_score(knn_pred, y_test)
print("Test accuracy: {:.2f}%".format(knn_acc*100))
```

```python
print(classification_report(y_test, logreg_pred))
```

<a id="rf"></a>
<h2><b><span style="color:#009b2d">üç∑ Step 4.6 | </span><span style="color:#9b006e">GaussianNB</span></b></h2>  

```python
gnb = GaussianNB()
gnb.fit(X_train, y_train)
gnb_pred = gnb.predict(X_test)
gnb_acc = accuracy_score(gnb_pred, y_test)
print("Test accuracy: {:.2f}%".format(gnb_acc*100))
```

```python
print(classification_report(y_test, logreg_pred))
```

<h2><b><span style="color:#009b2d">üç∑ Step 4.7 | </span><span style="color:#9b006e">Result</span></b></h2>
